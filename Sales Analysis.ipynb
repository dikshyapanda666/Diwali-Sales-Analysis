#Diwali Sales Analysis

**This project comes under the retail/e-commerce domain where we analyze Diwali sales data to understand customer behavior, purchasing patterns, and sales performance.”**

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('/content/Diwali Sales Data.csv', encoding='latin1')
```

**Number of Order Gender vise**

```python
sns.countplot(x='Gender', data=df)
```

```python
# From the above charts we have analyzed that the number of female in the data is more than number of males
```

**Age vise distribution**

```python
sns.countplot(data=df,x='Age Group',hue='Gender')
```

```python
# From the above chart we analyzed that the age group between 26-35 is more
```

**Sum of orders by Occupation**

```python
occ_sales = df.groupby('Occupation')['Orders'].sum().sort_values(ascending=False)

sns.barplot(x=occ_sales.values, y=occ_sales.index)
plt.title("Occupation-wise Sales")
```

```python
# from the above chart we analyzed that the maximum orders are from IT sector
```

**Total Number of count by marital status**

```python
sns.countplot(x='Marital_Status', data=df)
plt.title("Marital Status Distribution")
```

```python
# This chart analyzed that the married person can order more then unmerried
```

**Total  Persent of orders from top 5 states**

```python
state_sales = df.groupby('State')['Amount'].sum().sort_values(ascending=False).head(5)
plt.pie(state_sales, labels=state_sales.index, autopct='%1.1f%%', startangle=90)
plt.title("Top 5 States by Sales Amount")
```

**Average spending by Age**

```python
avg_age = df.groupby('Age')['Amount'].mean().reset_index()
sns.lineplot(x='Age', y='Amount', data=avg_age)
plt.title("Average Spending by Age")
```

```python
# from above charts we analyzed the which age people spends more
```

```python
sns.boxplot(x=df['Amount'])
plt.title("Sales Amount Boxplot")
```
